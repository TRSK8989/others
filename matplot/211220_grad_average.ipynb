```python
import csv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#経路指定
route_x = [60, 60, 30, 30, 0, 0, 7.5, 7.5, 15, 15] #x座標
route_y = [70, 0, 0, 70, 70, 0, 0, 65, 65, 0] #y座標


x = []
y = []
z = []

with open("test_1124.csv") as f:
    reader = csv.reader(f)
    for row in [row for row in reader]:
        x.append(row[0])
        y.append(float(row[1]))
        z.append(row[2])
x = [float(r) for r in x[1:]]
x.insert(0, 0.0)

up_nd = 0
fig_not_dot = []

for xC in x:

    udxC = str(xC).split(".")[1]
    fig_not_dot.append(udxC)
    
nozero = sum(map(lambda i: int(i) > 0, fig_not_dot))

if nozero > 1:
    up_nd = max([len(lang) for lang in fig_not_dot])


fig_not_dot = []
for yC in y:

    udyC = str(yC).split(".")[1]
    fig_not_dot.append(udyC)
    
nozero = sum(map(lambda i: int(i) > 0, fig_not_dot))

if nozero > 1:
    if max([len(lang) for lang in fig_not_dot]) > up_nd:
        up_nd = max([len(lang) for lang in fig_not_dot])


x_upnd = [int(r * (10**up_nd)) for r in x]
y_upnd = [int(r * (10**up_nd)) for r in y]




width = max(x_upnd)
height = max(y_upnd)
img = np.zeros([height + 1, width + 1])


pdx, pdxx, pdxy = [], [], []
pdy, pdyx, pdyy = [], [], []

for i in range(len(z)):
    img[y_upnd[i]][x_upnd[i]] = z[i]

    
for h in range(height + 1):
    
    ret_x = 0
    for w in range(width):
        if img[h][w] > 0:
            ret_x = img[h][w]
            dis = 1
        else:
            dis += 1
            
        if ret_x > 0 and img[h][w + 1] > 0:
            this_grad_x = (img[h][w + 1] - ret_x) / (dis / (10**up_nd))
            pdx.append(this_grad_x)
            pdxx.append((w + 1 - dis / 2) / (10**up_nd))
            pdxy.append(h / (10**up_nd))



for w in range(width + 1):
    
    ret_y = 0
    for h in range(height):
        if img[h][w] > 0:
            ret_y = img[h][w]
            dis = 1
        else:
            dis += 1
            
        if ret_y > 0 and img[h + 1][w] > 0:
            this_grad_y = (img[h + 1][w] - ret_y) / (dis / (10**up_nd))
            pdy.append(this_grad_y)
            pdyx.append(w / (10**up_nd))
            pdyy.append((h + 1 - dis / 2) / (10**up_nd))




x_dwnd = [x_ / (10**up_nd) for x_ in x_upnd]
y_dwnd = [y_ / (10**up_nd) for y_ in y_upnd]


off_route_pdx_flag = [1] * len(pdxx)
off_route_pdy_flag = [1] * len(pdyy)


for mv in range(len(route_x) - 1):
    x_high = max(route_x[mv + 1], route_x[mv])
    x_low = min(route_x[mv + 1], route_x[mv])
    y_high = max(route_y[mv + 1], route_y[mv])
    y_low = min(route_y[mv + 1], route_y[mv])

    if x_high - x_low > 0:
        y_base = route_y[mv]
        for i in range(len(pdxx)):
            if y_base - 0.01 < pdxy[i] and pdxy[i] < y_base + 0.01:
                if x_low - 0.01 < pdxx[i] and pdxx[i] < x_high + 0.01:
                    off_route_pdx_flag[i] = 0
                    
    elif y_high - y_low > 0:
        x_base = route_x[mv]
        for i in range(len(pdyy)):
            if x_base - 0.01 < pdyx[i] and pdyx[i] < x_base + 0.01:
                if y_low - 0.01 < pdyy[i] and pdyy[i] < y_high + 0.01:
                    off_route_pdy_flag[i] = 0

    
off_route_pdx = []
off_route_pdxx = []
off_route_pdxy = []

off_route_pdy = []
off_route_pdyx = []
off_route_pdyy = []
    
for n_flag in range(len(off_route_pdx_flag)):

    if off_route_pdx_flag[n_flag] == 1:
        off_route_pdx.append(abs(pdx[n_flag]))
        off_route_pdxx.append(pdxx[n_flag])
        off_route_pdxy.append(pdxy[n_flag])
        
for n_flag in range(len(off_route_pdy_flag)):

    if off_route_pdy_flag[n_flag] == 1:
        off_route_pdy.append(abs(pdx[n_flag]))
        off_route_pdyx.append(pdyx[n_flag])
        off_route_pdyy.append(pdyy[n_flag])
    
off_route_pdx_ave = sum(off_route_pdx) / len(off_route_pdx)
off_route_pdy_ave = sum(off_route_pdy) / len(off_route_pdy)
off_route_ave = (off_route_pdx_ave + off_route_pdy_ave) / 2


plt.figure(figsize=(16, 16))
plt.xlabel("X", fontsize=16)
plt.ylabel("Y", fontsize=16)
plt.grid()
plt.plot(x_dwnd, y_dwnd, linestyle='')
plt.plot(pdxx, pdxy, linestyle='', color='green')
plt.plot(pdyx, pdyy, linestyle='', color='green')


for i in range(len(z)):
    plt.text(x_dwnd[i], y_dwnd[i], str(z[i]), fontsize=12, color="blue", horizontalalignment='center', verticalalignment='center')


for i in range(len(off_route_pdx)):
    plt.text(off_route_pdxx[i], off_route_pdxy[i], round(off_route_pdx[i] / off_route_ave, 3), fontsize=12, horizontalalignment='center', verticalalignment='center')

for i in range(len(off_route_pdy)):
    plt.text(off_route_pdyx[i], off_route_pdyy[i], round(off_route_pdy[i] / off_route_ave, 3), fontsize=12, horizontalalignment='center', verticalalignment='center')
```
