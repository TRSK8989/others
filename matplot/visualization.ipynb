```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import random


x = np.array(range(0, 71, 5))
y = np.array(sorted(range(0, 71, 5), reverse=False))

Y, X = np.meshgrid(y, x)
X = X.reshape((225))
Y = Y.reshape((225))

z = list(range(0, 225))
Z = np.array(z) #np.array("")
print(Z)

plt.figure(figsize=(9, 9))
plt.xlabel("X", fontsize=16)
plt.ylabel("Y", fontsize=16)
plt.grid()
plt.plot(X, Y, marker='o', linestyle='')
for i in range(225):
    plt.text(X[i], Y[i], str(Z[i]))
#plt.show()
plt.savefig(".png")
```
